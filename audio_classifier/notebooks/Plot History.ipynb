```python
import os
import matplotlib.pyplot as plt
import pandas as pd
```

```python
log_csvs = sorted(os.listdir('../logs'))
print(log_csvs)
```

```python
labels = ['Conv 1D', 'Conv 2D', 'LSTM']
colors = ['r', 'm', 'c']
```

```python
fig, ax = plt.subplots(1, 3, sharey=True, figsize=(16,5))

for i, (fn, label, c) in enumerate(zip(log_csvs, labels, colors)):
    csv_path = os.path.join('..', 'logs', fn)
    df = pd.read_csv(csv_path)
    ax[i].set_title(label, size=16)
    ax[i].plot(df.accuracy, color=c, label='train')
    ax[i].plot(df.val_accuracy, ls='--', color=c, label='validation')
    ax[i].legend(loc='upper left')
    ax[i].tick_params(axis='both', which='major', labelsize=12)
    ax[i].set_ylim([0,1.1])

fig.text(0.5, 0.02, 'Epochs', ha='center', size=14)
fig.text(0.08, 0.5, 'Accuracy', va='center', rotation='vertical', size=14)
plt.show()
```

```python
fig, ax = plt.subplots(1, 3, sharey=True, figsize=(16,5))

for i, (fn, label, c) in enumerate(zip(log_csvs, labels, colors)):
    csv_path = os.path.join('..', 'logs', fn)
    df = pd.read_csv(csv_path)
    ax[i].set_title(label, size=16)
    ax[i].plot(df.loss, color=c, label='train loss')
    ax[i].plot(df.val_loss, ls='--', color=c, label='validation loss')
    ax[i].legend(loc='upper left')
    ax[i].tick_params(axis='both', which='major', labelsize=12)
    ax[i].set_ylim([0,2])

fig.text(0.5, 0.02, 'Epochs', ha='center', size=14)
fig.text(0.08, 0.5, 'Loss', va='center', rotation='vertical', size=14)
plt.show()
```

```python
fig, ax = plt.subplots(3, 2, figsize=(15, 20))  # 3 rows, 2 columns, larger figure size

for i, (fn, label, c) in enumerate(zip(log_csvs, labels, colors)):
    csv_path = os.path.join('..', 'logs', fn)
    df = pd.read_csv(csv_path)

    # Plotting the loss (first column)
    ax[i, 0].set_title(f'{label} - Loss', size=16)
    ax[i, 0].plot(df.loss, color=c, label='train loss')
    ax[i, 0].plot(df.val_loss, ls='--', color=c, label='validation loss')
    ax[i, 0].set_ylim([0, 2])
    ax[i, 0].tick_params(axis='both', which='major', labelsize=12)
    ax[i, 0].legend(loc='upper left')
    ax[i, 0].set_xlabel('Epochs', size=14)  # X-axis label for each subplot
    ax[i, 0].set_ylabel('Loss', size=14)  # Y-axis label for each subplot

    # Plotting the accuracy (second column)
    ax[i, 1].set_title(f'{label} - Accuracy', size=16)
    ax[i, 1].plot(df.accuracy, color=c, label='train accuracy')
    ax[i, 1].plot(df.val_accuracy, ls='--', color=c, label='validation accuracy')
    ax[i, 1].set_ylim([0, 1.1])
    ax[i, 1].tick_params(axis='both', which='major', labelsize=12)
    ax[i, 1].legend(loc='upper left')
    ax[i, 1].set_xlabel('Epochs', size=14)  # X-axis label for each subplot
    ax[i, 1].set_ylabel('Accuracy', size=14)  # Y-axis label for each subplot

plt.tight_layout()
plt.show()
```
